# Basic Preprocessing of Raw Scraped Data

```python
import pandas as pd;
import numpy as np;
import seaborn as sns;
import matplotlib.pyplot as plt;
import numpy as np;

from functools import reduce
from tqdm import tqdm

%matplotlib inline
```

## Import Raw Scraped Datasets

```python
df_2009 = pd.read_csv('raw_data/raw_data_2009.csv')
df_2010 = pd.read_csv('raw_data/raw_data_2010.csv')
df_2011 = pd.read_csv('raw_data/raw_data_2011.csv')
df_2012 = pd.read_csv('raw_data/raw_data_2012.csv')
df_2013 = pd.read_csv('raw_data/raw_data_2013.csv')
df_2014 = pd.read_csv('raw_data/raw_data_2014.csv')
df_2015 = pd.read_csv('raw_data/raw_data_2015.csv')
df_2016 = pd.read_csv('raw_data/raw_data_2016.csv')
df_2017 = pd.read_csv('raw_data/raw_data_2017.csv')
```

## Add Year Label and Merge 

```python
df_2009['Year'] = 2009
df_2010['Year'] = 2010
df_2011['Year'] = 2011
df_2012['Year'] = 2012
df_2013['Year'] = 2013
df_2014['Year'] = 2014
df_2015['Year'] = 2015
df_2016['Year'] = 2016
df_2017['Year'] = 2017
```

```python
dfs = [df_2009, df_2010, df_2011, df_2012, df_2013, df_2014, df_2015, df_2016, df_2017]

df = reduce(lambda left, right: pd.merge(left, right, how='outer'), dfs)
```

```python
df.head()
```

```python
df.info()
```

## Convert Event Date Location to a Simple Weekday Consistent Between All Years

```python
days = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday2', 'Monday2']

for day in days:
    df[day] = 0
```

```python
df.head()
```

```python
for i in tqdm(range(len(df))):

    dates = [date for date in df['Date and Time(s)'].values[i].split(' <br/>') if len(date) > 1]

    [date.split(' - ') for date in dates]

    year_string = str(df['Year'].values[i])

    if year_string in dates[0]:
        date_pairs = [date.split(', ' + year_string + ' ') for date in dates]
    else:
        date_pairs = [date.split(' - ') for date in dates]

    for date in date_pairs:
        if   'Sunday' in date[0] and 'August' in date[0]:
            df.loc[i, ('Sunday')]    = date[1]
        elif 'Monday' in date[0] and 'August' in date[0]:
            df.loc[i, ('Monday')]    = date[1]
        elif 'Tuesday' in date[0]:
            df.loc[i, ('Tuesday')]   = date[1]
        elif 'Wednesday' in date[0]:
            df.loc[i, ('Wednesday')] = date[1]
        elif 'Thursday' in date[0]:
            df.loc[i, ('Thursday')]  = date[1]
        elif 'Friday' in date[0]:
            df.loc[i, ('Friday')]    = date[1]
        elif 'Saturday' in date[0]:
            df.loc[i, ('Saturday')]  = date[1]
        elif 'Sunday' in date[0] and 'September' in date[0]:
            df.loc[i, ('Sunday2')]   = date[1]
        elif 'Monday' in date[0] and 'September' in date[0]:
            df.loc[i, ('Monday2')]   = date[1]
```

```python
df = df.drop(['Date and Time(s)'], axis=1)
```
